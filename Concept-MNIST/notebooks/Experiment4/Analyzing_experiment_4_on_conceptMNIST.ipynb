```python
import pandas as pd

df = pd.read_csv('concept-MNIST.csv')
df
```

```python
import matplotlib.pyplot as plt
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['num_of_new_concepts'], df['Sequential model accuracy on adversarial images'], label='Sequential Model', marker='o')
plt.plot(df['num_of_new_concepts'], df['Joint model accuracy on adversarial images'], label='Joint Model', marker='o')
plt.plot(df['num_of_new_concepts'], df['CNN model accuracy on adversarial images'], label='CNN Model', marker='o')

plt.xlabel('Number of New Concepts')
plt.ylabel('Accuracy on Adversarial Images')
plt.title('Model Accuracy on Adversarial Images vs Number of Concepts')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame

# 1. Accuracy Comparison on Normal vs Adversarial Images for Each Model
plt.figure(figsize=(12, 6))
models = ['Sequential', 'Joint', 'CNN']
for model in models:
    plt.plot(df['num_of_new_concepts'], df[f'{model} model accuracy on normal images'], label=f'{model} Normal', marker='o')
    plt.plot(df['num_of_new_concepts'], df[f'{model} model accuracy on adversarial images'], label=f'{model} Adversarial', linestyle='--', marker='o')
plt.xlabel('Number of New Concepts')
plt.ylabel('Accuracy')
plt.title('Model Accuracy on Normal vs Adversarial Images')
plt.legend()
plt.grid(True)
plt.show()

# 2. Accuracy Trend for Each Model with Increasing Number of Concepts
plt.figure(figsize=(12, 6))
for model in models:
    plt.plot(df['num_of_new_concepts'], df[f'{model} model accuracy on normal images'], label=f'{model}', marker='o')
plt.xlabel('Number of New Concepts')
plt.ylabel('Accuracy on Normal Images')
plt.title('Accuracy Trend with Increasing Number of Concepts')
plt.legend()
plt.grid(True)
plt.show()

# 3. Adversarial Training Impact
plt.figure(figsize=(12, 6))
for model in models:
    plt.plot(df['num_of_new_concepts'], df[f'{model} model accuracy on adversarial images'], label=f'{model} Pre-Training', marker='o')
    plt.plot(df['num_of_new_concepts'], df[f'{model} model accuracy on adversarial images after adversarial training'], label=f'{model} Post-Training', linestyle='--', marker='o')
plt.xlabel('Number of New Concepts')
plt.ylabel('Accuracy on Adversarial Images')
plt.title('Impact of Adversarial Training on Model Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
